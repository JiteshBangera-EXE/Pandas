Series= A pandas 1D labeled array  that can hold any data type (Think of it like a single column in a spreadsheet)

```python
data = [110, 130, 150, 170, 180, 200]  #creating a list of data

series = pd.Series(data, index=["a", "b", "c", "d", "e", "f"])  #creating a pandas series with the data and custom index

print(series)  #print the series

series.loc["c"] = 130  #unpdated the value of index "c" to 130
print(series)    #print the updated series

print(series.loc["b"])  #accessing the value of index "b" which is 120

print(series.iloc[0])  #accessing the value of index 0 which is 100

```

```python
print(series[series>150])  #filtering the series to get values greater than 150
print(series[series<150])  #filtering the series to get values less than 150
```

```python
calories = {"Day1": 1800, "Day2": 1750, "Day3": 2300}  #creating a dictionary of calories consumed on different days
calories_series = pd.Series(calories)  #creating a pandas series from the dictionary
print(calories_series)  #print the calories series

calories_series.loc["Day1"] += 500 #adding 500 calories to the value of "Day1"
print(calories_series)  #print the updated calories series
```

Dataframe= A tabular data structure which has rows and columns 2D

```python
data = {"Name": ["Alice", "Bob", "Charlie"],  
        "Age": [25, 30, 35]
    }   #creating a dictionary of data for a DataFrame

df = pd.DataFrame(data)  #creating a pandas DataFrame from the dictionary

print(df)  #print the DataFrame

df["Job"] = ["Engineer", "Doctor", "Artist"]  #adding a new column "Job" to the DataFrame

new_rows = pd.DataFrame([{"Name": "David", "Age": 40, "Job": "Lawyer"},
                         {"Name": "Kadu", "Age": 21, "Job": "Editor"}])  #creating a new DataFrame for the new rows

df = pd.concat([df, new_rows], ignore_index=True)  #concatenating the new row to the existing DataFrame
print(df)  #print the DataFrame
```

Filtering= Keeping a row that matches a condition

```python
tall_pokemon = df[df["Height"] >= 2.0]  #filtering the DataFrame to get Pokémon with height greater than and equal to 2.0
print(tall_pokemon)  #print the filtered DataFrame of tall Pokémon
```

```python
heavy_pokemon = df[df["Weight"] >= 200.0]  #filtering the DataFrame to get Pokémon with weight greater than and equal to 200.0
print(heavy_pokemon)  #print the filtered DataFrame of heavy Pokémon    
```

```python
water_pokemon = df[(df["Type1"] == "Water") | (df["Type2"] == "Water")]  #filtering the DataFrame to get Pokémon that are of type "Water" in either "Type1" or "Type2"
print(water_pokemon)  #print the filtered DataFrame of water Pokémon    
```

```python
ff_pokemon = df[(df["Type1"] == "Fire") & (df["Type2"] == "Flying")]  #filtering the DataFrame to get Pokémon that are of type "Fire" in "Type1" and "Flying" in "Type2"
print(ff_pokemon)  #print the filtered DataFrame of Fire and Flying type Pokémon    
```

Aggregate Function = Reduces a set of values into a single summary value 
                     Used to summarize and analyze data
                     Often used with the groupby() function

```python
# For Whole columns

mean_height = df.mean(numeric_only=True)
print(mean_height)  #print the mean height, weight, and Legendary status of all Pokémon

print(df.sum(numeric_only=True))  #print the sum of height, weight, and Legendary status of all Pokémon

print(df.min(numeric_only=True))  #print the minimum height, weight, and Legendary status of all Pokémon
print(df.max(numeric_only=True))  #print the maximum height, weight, and Legendary status of all Pokémon
print(df.count())  #print the count of non-null values in each column of the DataFrame
```

```python
#For Specific columns
print(df["Height"].mean())  #print the mean height of all Pokémon
print(df["Weight"].sum())  #print the sum of weight of all Pokémon
print(df["Height"].min())  #print the minimum height of all Pokémon
print(df["Weight"].max())  #print the maximum weight of all Pokémon
print(df["Type1"].value_counts())  #print the count of each unique value in the "Type1" column of the DataFrame
```

```python
# GroupBy()

group = df.groupby("Type1")  #grouping the DataFrame by the "Type1" column
print(group["Height"].mean())  #print the mean height of Pokémon for each unique value in the "Type1" column of the DataFrame
```

```python
print(group["Height"].count())  #print the count of Pokémon for each unique value in the "Type1" column of the DataFrame
```

Data CLeaning 

```python
# Drop irrelevant columns
print(df)  #print the column names of the DataFrame before dropping irrelevant columns
df = df.drop(columns=["Height"])  #dropping the "Type2" and "Legendary" columns from the DataFrame
print(df)  #print the updated DataFrame with irrelevant columns dropped
```

```python
# Handling missing values
df = df.fillna({"Type2" : "Nahi Hai"})  #filling any missing values in the DataFrame with 0
print(df.to_string())  #print the updated DataFrame with missing values handled

# df = df.dropna(subset=["Type2"])  #dropping any rows with missing values from the DataFrame
# print(df.to_string())  #print the updated DataFrame with rows containing missing values dropped
```

```python
# Fixing inconsistent data
df["Type1"] = df["Type1"].replace({"Grass": "GRASS",
                                   "Fire": "AAGGG"})  #replacing any occurrences of "Nahi Hai" in the "Type2" column with "None"
print(df.to_string())  #print the updated DataFrame with inconsistent data fixed
```

```python
# fix data types

df["Legendary"] = df["Legendary"].astype(bool)  #converting the "Legendary" column to boolean data type
print(df)  #print the data types of each column in the DataFrame
```
